```python
import pandas as pd
import os
import numpy as np
import pickle
```

### Load data from .txt files

```python
x = []
gyro = []
label = []
```

```python
num_sources = 7
num_digits = 10
num_files = 80
num_skips = 5
```

```python
for source in range(num_sources):
    for digit in range(num_digits):
        for i in range(num_files):
            #file_path = "/tmp/zfiles/training/{0}-{1}/accelerometer.txt".format(digit, i+1)
            #file_path = "../pixel3_data/test/kiel/{0}-{1}/accelerometer.txt".format(digit, i+1)
            file_path = "../pixel3_data/training/train_{2}/{0}-{1}/accelerometer.txt".format(digit, i+1, source+1)
            #print("The file path is {}".format(file_path))

            if os.path.exists(file_path):
                #print("The file exists.")
                file_obj = open(file_path, 'r')

                # skip the first 5 lines
                for _ in range(num_skips):
                    file_obj.readline()

                data_df = pd.read_csv(file_obj, sep=' ')
                #x.append(data_df[['elapsed-time-system', 'elapsed-time-sensor', 'x', 'y', 'z']].values)
                x.append(data_df[['x', 'y', 'z']].values)
                label.append(digit)
                #print('the sample size is {}'.format(x[-1].shape))
            else:
                pass
                #print("The file doesn't exists")
    print("Num of samples: {}".format(len(x)))
```

#### Accelerometer + Gyro
* The accelerometer files and gyro files don't match...

```python
x = []
label = []
```

```python
num_sources = 7
num_digits = 10
num_files = 100
num_skips = 5
```

```python
for source in range(num_sources):
    for digit in range(num_digits):
        for i in range(num_files):
            file_path = "../pixel3_data/training/{2}/{0}-{1}/accelerometer.txt".format(digit, i+1, source)
            file_path_gyro = "../pixel3_data/training/{2}/{0}-{1}/gyroscope-calibrated.txt".format(digit, i+1, source)
            #print("The file path is {}".format(file_path))

            if os.path.exists(file_path):
                #print("The file exists.")
                file_obj = open(file_path, 'r')
                file_obj_gyro = open(file_path_gyro, 'r')

                # skip the first 5 lines
                for _ in range(num_skips):
                    file_obj.readline()
                for _ in range(num_skips):
                    file_obj_gyro.readline()

                data_df = pd.read_csv(file_obj, sep=' ')
                data_df_gyro = pd.read_csv(file_obj_gyro, sep=' ')
                if data_df.shape[0] == data_df_gyro.shape[0]:
                    x.append(np.hstack([data_df[['x', 'y', 'z']].values, 
                                        data_df_gyro[['x', 'y', 'z']].values]))
                    #print('the sample size is {}'.format(x[-1].shape))
                    label.append(digit)
                else:
                    pass
                    #print("Accelerometer and Gyro mismatch!")
            else:
                pass
                #print("The file doesn't exists")
```
